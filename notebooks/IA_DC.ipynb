### Documentación del Uso de Inteligencia Artificial
##### Este proyecto ha integrado el uso responsable de herramientas de Inteligencia Artificial para potenciar el análisis de datos, la generación de documentación técnica y la resolución de problemas de programación.

---

## Herramientas Utilizadas y Tareas Específicas
---

#### Gemini (Google): Utilizada principalmente para la interpretación ejecutiva de los gráficos estadísticos, redacción de conclusiones del Análisis Exploratorio de Datos (EDA) e investigación de como generar el API.
---

#### DeepSeek: Empleada para la optimización de algoritmos de redes neuronales (ANN) y la corrección de errores específicos en el código.
---

#### ChatGPT (OpenAI): Utilizada para  evacuar dudas sobre limpieza, orden y funcionamiento general. Se pidió recomendación de como distribuir tareas en el equipo.
---

### Reflexión sobre el Aprendizaje
---
El uso de la IA en este proyecto transformó el proceso de desarrollo de una tarea puramente técnica a una de análisis crítico. En lugar de dedicar horas a la sintaxis repetitiva de las librerías de visualización o despliegue, pudimos enfocarnos en entender qué significaban realmente los datos para el negocio bancario. La IA actuó como un mentor disponible 24/7 que nos permitió iterar rápidamente sobre diferentes arquitecturas de redes neuronales y comprender conceptos complejos de probabilidad y estadística de forma aplicada. Así como investigar y comprender herramientas y funciones las cuáles no habíamos manejado anteriormente.

Además, nos enseñó la importancia de la precisión en la comunicación: para obtener una buena respuesta de la IA, primero debíamos tener muy claro el problema técnico, lo que reforzó nuestro dominio sobre el flujo de trabajo de Machine Learning (desde el preprocesamiento hasta el despliegue).

### Qué modificaciones hicimos al código/análisis generado por IA
---

#### Adaptación a versiones actuales de librerías

Se modificaron instrucciones que ya no son compatibles con versiones recientes de TensorFlow / Keras.
Se eliminó model.predict_classes() (obsoleto)


---

Se sustituyó por:
predicciones = np.argmax(model.predict(X_test), axis=1)
Esto garantizó compatibilidad y evitó errores de ejecución.


#### Corrección de errores de tipo de dato
Se ajustaron parámetros que generaban errores por no cumplir el formato esperado por Keras.

---

np.round() devolvía valores float
Se forzó conversión a entero:
model.add(Dense(units=int(np.round(num_neuronas/2)), activation='relu'))


#### Reestructuración del flujo del EDA


Verificación de valores nulos antes de visualizaciones

Separación clara entre variables numéricas y categóricas

Orden lógico: distribución → correlación → relaciones con la clase

El uso de IA se empleó como herramienta de apoyo para acelerar la construcción del código, pero el análisis, validación de errores, ajustes técnicos e interpretación estadística fueron realizados manualmente para asegurar la validez del estudio.
