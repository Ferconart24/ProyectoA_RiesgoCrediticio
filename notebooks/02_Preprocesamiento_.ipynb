```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Importar configuración y funciones del proyecto
import sys
sys.path.append('..')
from src import config
from src.data_prep import load_data, preprocess_pipeline, split_data

print("Módulos importados correctamente.")
```

```python
# Cargar el dataset original (german_credit.csv)
df = load_data()  # esto lee config.RAW_DATA_FILE

print(f"Dataset cargado: {df.shape}")
print("\nPrimeras filas:")
df.head()
```

```python
# ============================================================
# PREPROCESAMIENTO PARA CLASIFICACIÓN BINARIA
# ============================================================
print("="*60)
print("PROCESANDO PARA CLASIFICACIÓN BINARIA")
print("="*60)

X_bin, y_bin, scaler_bin, encoders_bin = preprocess_pipeline(
    df,
    fit=True,
    target_type='binary',
    save=True  # guarda scaler y encoders en models/
)

print(f"\nX_bin shape: {X_bin.shape}")
print(f"y_bin shape: {y_bin.shape}")
print("Distribución de y_bin:")
print(pd.Series(y_bin).value_counts())
```

```python
# Dividir en entrenamiento y prueba (binario)
X_train_bin, X_test_bin, y_train_bin, y_test_bin = split_data(
    X_bin, y_bin,
    test_size=config.TEST_SIZE,
    random_state=config.RANDOM_STATE
)

print("\nTamaños después de split (binario):")
print(f"X_train_bin: {X_train_bin.shape}")
print(f"X_test_bin: {X_test_bin.shape}")
print(f"y_train_bin: {y_train_bin.shape}")
print(f"y_test_bin: {y_test_bin.shape}")
```

```python
# ============================================================
# PREPROCESAMIENTO PARA CLASIFICACIÓN MULTICLASE
# ============================================================
print("\n" + "="*60)
print("PROCESANDO PARA CLASIFICACIÓN MULTICLASE")
print("="*60)

X_multi, y_multi, scaler_multi, encoders_multi = preprocess_pipeline(
    df,
    fit=True,
    target_type='multiclass',
    save=True
)

print(f"\nX_multi shape: {X_multi.shape}")
print(f"y_multi shape: {y_multi.shape}")
print("Distribución de y_multi:")
print(pd.Series(y_multi).value_counts().sort_index())
```

```python
# Dividir en entrenamiento y prueba (multiclase)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = split_data(
    X_multi, y_multi,
    test_size=config.TEST_SIZE,
    random_state=config.RANDOM_STATE
)

print("\nTamaños después de split (multiclase):")
print(f"X_train_multi: {X_train_multi.shape}")
print(f"X_test_multi: {X_test_multi.shape}")
print(f"y_train_multi: {y_train_multi.shape}")
print(f"y_test_multi: {y_test_multi.shape}")
```

```python
# ============================================================
# GUARDAR DATOS PROCESADOS EN DIFERENTES FORMATOS
# ============================================================
print("\n" + "="*60)
print("GUARDANDO DATOS PROCESADOS")
print("="*60)

# Guardar arrays numpy (formato eficiente para TensorFlow)
np.save(config.PROCESSED_DATA_DIR / 'X_train_bin.npy', X_train_bin)
np.save(config.PROCESSED_DATA_DIR / 'X_test_bin.npy', X_test_bin)
np.save(config.PROCESSED_DATA_DIR / 'y_train_bin.npy', y_train_bin)
np.save(config.PROCESSED_DATA_DIR / 'y_test_bin.npy', y_test_bin)

np.save(config.PROCESSED_DATA_DIR / 'X_train_multi.npy', X_train_multi)
np.save(config.PROCESSED_DATA_DIR / 'X_test_multi.npy', X_test_multi)
np.save(config.PROCESSED_DATA_DIR / 'y_train_multi.npy', y_train_multi)
np.save(config.PROCESSED_DATA_DIR / 'y_test_multi.npy', y_test_multi)

print("Archivos .npy guardados en:", config.PROCESSED_DATA_DIR)
```

```python
# Guardar también en formato CSV (para revisión fácil)
# Para binario
df_train_bin = pd.DataFrame(X_train_bin)
df_train_bin['target'] = y_train_bin
df_test_bin = pd.DataFrame(X_test_bin)
df_test_bin['target'] = y_test_bin

df_train_bin.to_csv(config.PROCESSED_DATA_DIR / 'train_binary.csv', index=False)
df_test_bin.to_csv(config.PROCESSED_DATA_DIR / 'test_binary.csv', index=False)

# Para multiclase
df_train_multi = pd.DataFrame(X_train_multi)
df_train_multi['risk_level'] = y_train_multi
df_test_multi = pd.DataFrame(X_test_multi)
df_test_multi['risk_level'] = y_test_multi

df_train_multi.to_csv(config.PROCESSED_DATA_DIR / 'train_multiclass.csv', index=False)
df_test_multi.to_csv(config.PROCESSED_DATA_DIR / 'test_multiclass.csv', index=False)

print("\nArchivos CSV guardados:")
print(f"    train_binary.csv: {df_train_bin.shape}")
print(f"    test_binary.csv: {df_test_bin.shape}")
print(f"    train_multiclass.csv: {df_train_multi.shape}")
print(f"    test_multiclass.csv: {df_test_multi.shape}")
```

```python
# ============================================================
# VERIFICACIONES FINALES
# ============================================================
print("\n" + "="*60)
print("VERIFICACIÓN FINAL")
print("="*60)

print("Binario:")
print(f"  X_train: {X_train_bin.shape}, y_train: {y_train_bin.shape}")
print(f"  X_test: {X_test_bin.shape}, y_test: {y_test_bin.shape}")
print("\nMulticlase:")
print(f"  X_train: {X_train_multi.shape}, y_train: {y_train_multi.shape}")
print(f"  X_test: {X_test_multi.shape}, y_test: {y_test_multi.shape}")

# Verificar consistencia de dimensiones
assert X_train_bin.shape[0] == y_train_bin.shape[0], "Error en binario train"
assert X_test_bin.shape[0] == y_test_bin.shape[0], "Error en binario test"
assert X_train_multi.shape[0] == y_train_multi.shape[0], "Error en multi train"
assert X_test_multi.shape[0] == y_test_multi.shape[0], "Error en multi test"

print("\n Todas las dimensiones son correctas.")
print("\n Los datos están listos para ser usados en los notebooks de modelos.")
```
